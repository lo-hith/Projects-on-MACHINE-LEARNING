```python
# This are the libraries we have to inport 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#this defines as S.No is no use in data set so we are droping this column
data.drop(["S. No."],axis=1,inplace=True)
```

```python
data
```

```python
#In this we are renaming the colums 
data.rename(columns={"Name of State / UT":"STATE","Cured/Discharged/Migrated":"DISCHARGED"},inplace=True)
```

```python
data.head(5)
```

```python
#Plotting regarding the Total Active Cases by State
plt.figure(figsize=(20,10))
plt.bar(data["STATE"],data["Active Cases"],edgecolor="black")
plt.title("Total Active Cases By State",fontsize=18)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Cases",fontsize=14)
plt.xticks(rotation=90,size=12)
plt.yticks(size=12)
plt.show()
```

```python
#Plotting regarding the Total DISCHARGED PATIENTS  by State wise
plt.figure(figsize=(20,10))
plt.bar(data["STATE"],data["DISCHARGED"],edgecolor="black")
plt.title("Total Discharged Patients",fontsize=16)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Patients",fontsize=14)
plt.xticks(rotation=90,size=15)
plt.yticks(size=15)
plt.show()
```

```python
#Plotting regarding the Total Death Cases by State wise
plt.figure(figsize=(18,10))
plt.bar(data["STATE"],data["Deaths"],edgecolor="brown")
plt.title("Total Deaths",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Patients",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.bar(data["STATE"],data["Total Confirmed cases"],edgecolor="black")
plt.title("Total Confirmed cases ",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Patients",fontsize=14)
plt.xticks(rotation=90,size=16)
plt.yticks(size=16)
plt.show()
```
